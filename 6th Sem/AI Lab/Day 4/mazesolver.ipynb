```python
from collections import deque
print("Nipun Kumar Ranjan")
print("23052738")
print("CSE-20")
maze = [
    ['A', 'B', 'C', 'D', 'M'],
    ['0', '0', '0', 'E', 'N'],
    ['K', 'J', 'I', 'F', 'O'],
    ['L', '0', '0', 'G', 'P'],
    ['M', 'N', 'O', 'H', 'Q']
]
start = (0, 0)  # 'A'
end = (3, 4)    # 'P'

def bfs_maze(maze, start, end):
    rows, cols = len(maze), len(maze[0])
    queue = deque([(start, [start])])
    visited = {start}
    nodes_explored = 0
    
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    
    while queue:
        (row, col), path = queue.popleft()
        nodes_explored += 1
        
        if (row, col) == end:
            return path, nodes_explored
        
        for dr, dc in directions:
            new_row, new_col = row + dr, col + dc
            
            if (0 <= new_row < rows and 0 <= new_col < cols and 
                (new_row, new_col) not in visited and maze[new_row][new_col] != '0'):
                
                visited.add((new_row, new_col))
                queue.append(((new_row, new_col), path + [(new_row, new_col)]))
    
    return None, nodes_explored

def dfs_maze(maze, start, end):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    nodes_explored = [0]
    
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    
    def dfs(row, col, path):
        nodes_explored[0] += 1
        
        if (row, col) == end:
            return path
        
        visited.add((row, col))
        
        for dr, dc in directions:
            new_row, new_col = row + dr, col + dc
            
            if (0 <= new_row < rows and 0 <= new_col < cols and 
                (new_row, new_col) not in visited and maze[new_row][new_col] != '0'):
                
                result = dfs(new_row, new_col, path + [(new_row, new_col)])
                if result:
                    return result
        
        return None
    
    path = dfs(start[0], start[1], [start])
    return path, nodes_explored[0]

# BFS Results
bfs_path, bfs_nodes = bfs_maze(maze, start, end)
print("BFS Results:")
print(f"Shortest path: {bfs_path}")
print(f"Nodes explored: {bfs_nodes}\n")

# DFS Results
dfs_path, dfs_nodes = dfs_maze(maze, start, end)
print("DFS Results:")
print(f"Valid path: {dfs_path}")
print(f"Nodes explored: {dfs_nodes}\n")

# Comparison
print("Comparison:")
print(f"BFS explored {bfs_nodes} nodes, DFS explored {dfs_nodes} nodes")
print(f"BFS path length: {len(bfs_path) if bfs_path else 0}")
print(f"DFS path length: {len(dfs_path) if dfs_path else 0}")
```
