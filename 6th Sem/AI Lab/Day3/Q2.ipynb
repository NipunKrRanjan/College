```python
import random
def generate_binary_matrix_simple(m, n, p_one=0.7, seed=42):
    rng = random.Random(seed)
    matrix = []
    for i in range(m):
        row = []
        for j in range(n):
            random_value = rng.random()
            if random_value < p_one:
                row.append(1)
            else:
                row.append(0)
        matrix.append(row)   
    return matrix
m=int(input("Enter the number of rows"))
n=int(input("Enter number of columns"))
matrix = generate_binary_matrix_simple(m, n, p_one=0.7, seed=42)
print("Matrix Generated")
for row in matrix:
    print(row)
def find_paths(matrix, n):
    paths = []
    if matrix[0][0] == 0 or matrix[n-1][n-1] == 0:
        return paths
    def dfs(i, j, path):
        if i == n-1 and j == n-1:
            paths.append(path.copy())
            return
        if j+1 < n and matrix[i][j+1] == 1: 
            path.append((i, j+1))
            dfs(i, j+1, path)
            path.pop()
        if i+1 < n and matrix[i+1][j] == 1:
            path.append((i+1, j))
            dfs(i+1, j, path)
            path.pop()
    dfs(0, 0, [(0, 0)])
    return paths

print("\nGenerated Matrix:")
for row in matrix:
    print(" ".join(map(str, row)))

all_paths = find_paths(matrix, n)
print("\nPaths found:")
if not all_paths:
    print("No path found")
else:
    for pth in all_paths:
        print(", ".join(f"({r},{c})" for r, c in pth))
    print("Total paths:", len(all_paths))
```
