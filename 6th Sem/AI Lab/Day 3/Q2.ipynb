```python
import random


n = int(input("Enter size: "))
p = 0.7

matrix = [[1 if random.random() < p else 0 for _ in range(n)] for _ in range(n)]

sr = int(input("Enter start row: "))
sc = int(input("Enter start col: "))
er = int(input("Enter end row: "))
ec = int(input("Enter end col: "))

matrix[sr][sc] = 1
matrix[er][ec] = 1

def find_paths(matrix, n, sr, sc, er, ec):
    paths = []
    if matrix[sr][sc] == 0 or matrix[er][ec] == 0:
        return paths
    def dfs(i, j, path):
        if i == er and j == ec:
            paths.append(path.copy())
            return
        if j+1 < n and matrix[i][j+1] == 1:
            path.append((i, j+1))
            dfs(i, j+1, path)
            path.pop()
        if i+1 < n and matrix[i+1][j] == 1:
            path.append((i+1, j))
            dfs(i+1, j, path)
            path.pop()
    dfs(sr, sc, [(sr, sc)])
    return paths

print("\nGenerated Matrix:")
for row in matrix:
    print(" ".join(map(str, row)))

all_paths = find_paths(matrix, n, sr, sc, er, ec)
print("\nPaths found:")

if not all_paths:
    print("No path found")
else:
    for pth in all_paths:
        print(", ".join(f"({r},{c})" for r, c in pth))
    print("Total paths:", len(all_paths))
```
