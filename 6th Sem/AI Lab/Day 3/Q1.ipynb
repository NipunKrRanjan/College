```python
import random
def generate_binary_matrix_simple(m, n, p_one=0.7, seed=42):
    rng = random.Random(seed)
    matrix = []
    for i in range(m):
        row = []
        for j in range(n):
            random_value = rng.random()
            if random_value < p_one:
                row.append(1)
            else:
                row.append(0)
        matrix.append(row)   
    return matrix
m=int(input("Enter the number of rows"))
n=int(input("Enter number of columns"))
matrix = generate_binary_matrix_simple(m, n, p_one=0.7, seed=42)
for row in matrix:
    print(row)
```
