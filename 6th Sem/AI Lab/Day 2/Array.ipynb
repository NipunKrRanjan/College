```python
print("Nipun Kumar Ranjan\n23052738\n")
print("Array Creation")
fruits = ["Apple", "Banana", "Cherry", "Date"]
print("Original array:", fruits)
print("\n")
print("Array Element Access")
print("First element:", fruits[0])
print("Last element:", fruits[-1])

print("\n")
print("Array Length")
print("Array length:", len(fruits))

print("\nLooping Array Elements")
print("All elements:")
for fruit in fruits:
    print(fruit)

print("\nAdding Array Elements using append()")
fruits.append("Elderberry")
print("After append:", fruits)

print("\nRemoving array elements using pop()")
removed = fruits.pop()
print("Removed element:", removed)
print("Final array:", fruits)
```
