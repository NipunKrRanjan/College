```python
# I. Creation, access, update
d = {'a': 1, 'b': 2}
print(d['a'])        # access
d['c'] = 3           # add
d['a'] = 10          # update
print(d)

# II. length, type, dict() constructor
e = dict(x=100, y=200)
print(len(d), type(d), e)

# III. add/remove (pop, popitem) and checking
d['d'] = 4
print('before pop:', d)
p = d.pop('b')       # remove by key
print('popped b ->', p, 'now', d)
k, v = d.popitem()   # remove last item
print('popitem ->', k, v, 'now', d)
print('c in d?', 'c' in d)

# IV. clear() and del()
f = d.copy()
f.clear()
print('after clear f:', f)
del f
try:
    print(f)
except NameError:
    print('f is deleted')

# V. looping and copying
g = {'k1': 1, 'k2': 2}
h = g.copy()
for key, val in g.items():
    print(key, val)
print('copy h:', h)

# VI. nested dictionary - create, access, loop
nested = {
    'p1': {'name': 'Alice', 'age': 25},
    'p2': {'name': 'Bob',   'age': 30}
}
print(nested['p1']['name'])          # access nested
for pid, info in nested.items():
    print(pid, info['name'], info['age'])
```
