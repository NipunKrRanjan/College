```python
import heapq
print("Name: Nipun Kumar Ranjan\n23052738\nCSE-20")
def get_manhattan(board):
    distance = 0
    for i, tile in enumerate(board):
        if tile != 0:
            r, c = divmod(i, 3)
            gr, gc = divmod(tile - 1, 3)
            distance += abs(r - gr) + abs(c - gc)
    return distance

def solve(start, goal):
    queue = [(get_manhattan(start), start, [])]
    seen = {start: 0}

    while queue:
        _, board, path = heapq.heappop(queue)

        if board == goal:
            return path

        idx = board.index(0)
        r, c = divmod(idx, 3)

        for dr, dc, move in [(-1, 0, 'Up'), (1, 0, 'Down'), (0, -1, 'Left'), (0, 1, 'Right')]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < 3 and 0 <= nc < 3:
                neighbor = list(board)
                target = nr * 3 + nc
                neighbor[idx], neighbor[target] = neighbor[target], neighbor[idx]
                neighbor = tuple(neighbor)
                
                new_cost = len(path) + 1
                if neighbor not in seen or new_cost < seen[neighbor]:
                    seen[neighbor] = new_cost
                    f_score = new_cost + get_manhattan(neighbor)
                    heapq.heappush(queue, (f_score, neighbor, path + [move]))
    return None

start_node = (1, 2, 3, 0, 4, 6, 7, 5, 8)
goal_node = (1, 2, 3, 4, 5, 6, 7, 8, 0)

solution = solve(start_node, goal_node)
print(solution)
```
