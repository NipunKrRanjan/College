```python
import numpy as np
print("Name: Nipun Kumar Ranjan\n23052738\nCSE-20")
adjacency_matrix = np.array([
  # sq  6  3  MS  12  14  25  
    [0, 1, 0, 0,  0,  0,  0], #sq
    [1, 0, 1, 0,  1,  0,  0], #6
    [0, 1, 0, 1,  0,  0,  0], #3
    [0, 0, 1, 0,  0,  1,  0], #MS
    [0, 1, 0, 0,  0,  1,  1], #12
    [0, 0, 0, 1,  1,  0,  1], #14
    [0, 0, 0, 0,  1,  1,  0], #25
])

def bi_directional_bfs(start, end, adjacency_matrix):
    # Initialize the queues for both directions
    queue_start = [start]
    queue_end = [end]
    visited_start = {start}
    visited_end = {end}

    while queue_start and queue_end:
        # Expand from the start node
        current_start = queue_start.pop(0)
        for neighbor in range(len(adjacency_matrix[current_start])):
            if adjacency_matrix[current_start][neighbor] == 1 and neighbor not in visited_start:
                if neighbor in visited_end:
                    return True  # Path found
                visited_start.add(neighbor)
                queue_start.append(neighbor)

        # Expand from the end node
        current_end = queue_end.pop(0)
        for neighbor in range(len(adjacency_matrix[current_end])):
            if adjacency_matrix[current_end][neighbor] == 1 and neighbor not in visited_end:
                if neighbor in visited_start:
                    return True  # Path found
                visited_end.add(neighbor)
                queue_end.append(neighbor)

    return False  # No path found

# Example usage
start_node = (0, 0)
end_node = (6, 6)
result = bi_directional_bfs(start_node[0], end_node[0], adjacency_matrix)

# Print the path if it exists
if result:
    print(f"Path exists from {start_node} to {end_node}")
else:
    print("No path exists")
```
