```python

print("I. Set Creation and Set Length")
s1 = {1, 2, 3, 4}
print("s1:", s1, "len:", len(s1))

print("\nII. Allocation of Different Data Types into a Set")
mixed = {1, "two", 3.0, (4, 5)}
print("mixed:", mixed)

print("\nIII. Using set() constructor")
s_from_list = set([1, 2, 2, 3])
s_from_string = set("hello")
print("from list:", s_from_list)
print("from string (unique chars):", s_from_string)

print("\nIV. Accessing Set Items (membership & iteration)")
print("2 in s1?", 2 in s1)
print("Iterating s1:")
for item in s1:
    print(item, end=" ")
print()

print("\nV. Adding items to a Set using add() and update()")
s2 = {1, 2}
s2.add(3)
s2.update([4, 5], {6, 7})
print("s2 after add/update:", s2)

print("\nVI. Remove set item using remove(), pop() and discard()")
s3 = {10, 20, 30}
s3.remove(20)          # KeyError if missing
popped = s3.pop()      # removes an arbitrary element
print("after remove/pop ->", s3, "popped:", popped)
s3.discard(999)        # no error if missing
print("after discard(non-existent):", s3)

print("\nVII. Use clear() and del() on a set")
s4 = {1, 2, 3}
s4.clear()
print("s4 after clear():", s4)

print("\nVIII. Set Looping - for and while")
s_loop = {"a", "b", "c"}
print("for-loop:")
for x in s_loop:
    print(x, end=" ")
print("\nwhile-loop (popping until empty):")
s_temp = {1, 2, 3}
while s_temp:
    x = s_temp.pop()
    print("popped", x, "remaining", s_temp)

print("\nIX. Set Joining using union(), update(), intersection(), intersection_update()")
A = {1, 2, 3}
B = {3, 4, 5}
print("A | B (union):", A.union(B))
C = A.copy()
C.update(B)
print("A updated with B (update):", C)
print("A & B (intersection):", A.intersection(B))
D = A.copy()
D.intersection_update(B)
print("A after intersection_update with B:", D)

print("\nX. difference(), difference_update(), isdisjoint() and symmetric_difference()")
X = {1, 2, 3, 4}
Y = {3, 4, 5}
print("X - Y (difference):", X.difference(Y))
Z = X.copy()
Z.difference_update(Y)
print("X after difference_update(Y):", Z)
print("X and Y disjoint?", X.isdisjoint(Y))
print("X ^ Y (symmetric difference):", X.symmetric_difference(Y))
```
