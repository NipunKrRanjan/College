```python
my_tuple = (1, 5, 10, 'red', 'blue', 10, 20)

print("Tuple Creation")
print("INITIAL TUPLE")
print("Start Tuple: {my_tuple}\n")

print("Duplicate Tuple")
duplicate_tuple = my_tuple
print("Duplicated Tuple: ",{duplicate_tuple})
tuple_length = len(my_tuple)
print("Length: ",{tuple_length})

print("ACCESSING TUPLES")
print("Positive Index [0]:" ,{my_tuple[0]})
print("Negative Index [-1]:",{my_tuple[-1]})
print("Slice [1:4]:",{my_tuple[1:4]})
is_present = 'red' in my_tuple
print("Check ('red' in tuple):",{is_present})

print("MODIFYING (Add/Remove)")
temp_list = list(my_tuple)
print("Converted to List:",{temp_list})
temp_list.append(50)
temp_list.remove(10)
new_tuple = tuple(temp_list)
print("Modified and back to Tuple:", {new_tuple})

print("LOOPING (for loop)")
for item in my_tuple:
    print("Item found:",{item})

print("JOINING & MULTIPLICATION")
tuple_a = (1, 2)
tuple_b = ('x', 'y')
joined_tuple = tuple_a + tuple_b
print(f"Joining ({tuple_a} + {tuple_b}): {joined_tuple}")
multiplied_tuple = tuple_a * 3
print(f"Multiplication ({tuple_a} * 3): {multiplied_tuple}")

print("COUNT() & INDEX()")
count_result = my_tuple.count(10)
print(f"Count of 10 in {my_tuple}: {count_result}")
index_result = my_tuple.index('blue')
print(f"Index of 'blue' in {my_tuple}: {index_result}")
```
